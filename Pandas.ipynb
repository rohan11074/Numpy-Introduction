```python
import pandas as pd
Google_stock = pd.read_csv('./GOOG.csv')
Amazon_stock = pd.read_csv('./AMZN.csv')
Apple_stock = pd.read_csv('./AAPL.csv')

Google_stock = Google_stock.drop(['Open','High','Low','Close','Volume'], axis = 1)
Amazon_stock = Amazon_stock.drop(['Open','High','Low','Close','Volume'], axis = 1)
Apple_stock = Apple_stock.drop(['Open','High','Low','Close','Volume'], axis = 1)


Google_stock = Google_stock.set_index('Date')
Google_stock=Google_stock.rename(columns={'Adj Close':'G sock'})

Amazon_stock = Amazon_stock.set_index('Date')
Amazon_stock=Amazon_stock.rename(columns={'Adj Close':'A sock'})

Apple_stock = Apple_stock.set_index('Date')
Apple_stock=Apple_stock.rename(columns={'Adj Close':'Apple sock'})

dates = pd.date_range('2000-01-01', '2016-12-31')

# We create and empty DataFrame that uses the above dates as indices
all_stocks = pd.DataFrame(index = dates)
all_stocks = all_stocks.join(Google_stock)
all_stocks = all_stocks.join(Amazon_stock)
all_stocks = all_stocks.join(Apple_stock)
all_stocks
```
